```python
df=df.dropna(subset=['customerid','description','invoicedate'])
```

```python
df=df[df['quantity']>0 &(df['unitprice']>0)]
```

```python
df.info()
```

```python
df=df.drop_duplicates()
```

```python
df.info()
```

```python
df['country']=df['country'].str.lower().str.strip().str.title()
```

```python
df.columns
```

```python
df['country'].unique()
```

```python
df['description']=df['description'].str.strip()
```

```python
df['description'] = df['description'].str.replace(r'[^A-Za-z0-9 ]', '', regex=True)
```

```python
df['stockcode'] = df['stockcode'].astype(str)
df['customerid'] = df['customerid'].astype(int)
```

```python
df.info()
```

```python
df['invoiceno'] = df['invoiceno'].astype(str)
```

```python
df['invoiceno'].str.startswith('c')
```

```python
df = df[df['quantity'] < 10000]
df = df[df['unitprice'] < 10000]
```

```python
df.info()
```

```python
df = df[(df['invoicedate'].dt.year >= 2010) & (df['invoicedate'].dt.year <= 2012)]
```

```python
duplicates = df.duplicated(subset=['invoiceno','stockcode'], keep=False)
df = df[~duplicates]
```

```python
df.info()
```

```python
df['year']=df['invoicedate'].dt.year
```
